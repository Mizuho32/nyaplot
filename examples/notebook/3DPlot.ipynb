```ruby
require 'nyaplot'
require 'nyaplot3d'
```

```ruby
Nyaplot.init_iruby
```

```ruby
x=[];y=[];z=[]
-10.step(10, 1.0) do |i|
  -10.step(10, 1.0) do |j|
    x.push(i)
    y.push(j)
    z.push(Math.sin(Math.sqrt(i*i+j*j))/Math.sqrt(i*i+j*j))
  end
end
z.map!{|val| next (val.nan? ? 0 : val)}
print("")
```

```ruby
plot = Nyaplot::Plot3D.new
plot.add(:wireframe, x, y, z)
plot.show
```
